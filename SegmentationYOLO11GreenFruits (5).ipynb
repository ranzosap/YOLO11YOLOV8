```python
# Import necessary libraries
from ultralytics import YOLO
import cv2
import os

# Define paths
model_path = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/YOLO11instanceSegmentation/dataset/DatasetFINAL/OccludedNonOccluded/YOLOv8InstanceSegmentation/YOLOv8n/train/weights/last.pt'
image_directory = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/YOLO11instanceSegmentation/dataset/DatasetFINAL/OccludedNonOccluded/ImageAnalysis/Maain/ActualIMage'
output_directory = '/media/mzc/e4fcae9b-7400-4641-a800-034f249f8d14/mzc/YOLO11instanceSegmentation/dataset/DatasetFINAL/OccludedNonOccluded/ImageAnalysis/Maain/Yolov8Prediction'

# Load the YOLO11n model
model = YOLO(model_path)

# Ensure output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate over each image in the image directory
for image_file in os.listdir(image_directory):
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter for image files
        image_path = os.path.join(image_directory, image_file)
        img = cv2.imread(image_path)

        if img is not None:
            # Run inference on the image
            results = model(image_path)

            # Get the annotated image with predictions
            annotated_img = results[0].plot()

            # Save the annotated image to the output directory
            output_image_path = os.path.join(output_directory, f"predicted_{image_file}")
            cv2.imwrite(output_image_path, annotated_img)
            print(f"Saved predicted image: {output_image_path}")
        else:
            print(f"Failed to load image: {image_path}")

print("All predictions have been saved!")
```
